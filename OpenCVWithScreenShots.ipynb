```python
from mss import mss
import pyscreenshot as ImageGrab
import numpy as np
import cv2
import time
from PIL import ImageGrab
```

```python


im = ImageGrab.grab()
#PEGAND A TELA INTEIRA
im.save('Fullscreen.png')
```

```python


ultima = time.time()

while(True):
    #Um pedaço da tela inteira
    frame = np.array(ImageGrab.grab(bbox=(0, 0, 300, 300)))

    FPS = 1.0 / (time.time() - ultima)
    print(f'FPS {FPS}')
    ultima = time.time()

    cv2.imshow('Tela', cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    if cv2.waitKey(1) == 13:
        print('saindo..')
        break

cv2.destroyAllWindows()          
```

```python
#PRECISA DE MAIS FPS ? TENTE ISSO

bb = {'top': 100, 'left': 0, 'width': 400, 'height': 300}
sct = mss()
contador = 0
ultima = time.time()
while True:
    contador += 1

    sctImage = sct.grab(bb)

    cv2.imshow('Tela', np.array(sctImage))

    if contador % 30 == 0:
        FPS = 1.0/(time.time() - ultima)
        print(f'FPS {FPS}')

    ultima = time.time()

    if cv2.waitKey(1) == 13:
        print('saindo..')
        break

cv2.destroyAllWindows()
```
