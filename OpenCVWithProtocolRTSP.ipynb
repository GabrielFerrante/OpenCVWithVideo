# **RTSP Protocolo com OpenCV**

O RTSP (Real-Time Streaming Protocol)  um protocolo de rede utilizado para controlar a transmisso de contedo multimdia em tempo real, como udio e vdeo. Ele foi desenvolvido para estabelecer e gerenciar sesses entre clientes (ex.: players de vdeo) e servidores de mdia (ex.: cmeras IP, servidores de streaming), permitindo operaes como iniciar, pausar ou encerrar uma transmisso.

## Principais caractersticas:

Funo Principal:
Controlar a reproduo de fluxos de mdia em tempo real, mas no transmite os dados em si. O transporte dos dados (udio/vdeo) geralmente  feito por protocolos como RTP (Real-Time Transport Protocol) ou RTCP (Real-Time Control Protocol).

Mtodos Similares ao HTTP:
- Usa comandos de texto, como:

    - OPTIONS: Consulta recursos suportados pelo servidor.

    - DESCRIBE: Obtm metadados sobre a mdia (ex.: formato, codec).

    - SETUP: Configura a conexo para transporte dos dados.

    - PLAY: Inicia a transmisso.

    - PAUSE: Pausa temporariamente.

    - TEARDOWN: Encerra a sesso.

Portas e Transporte:
- Geralmente opera na porta 554 (TCP ou UDP). O RTSP pode usar TCP para controle e UDP para transporte de mdia (priorizando baixa latncia).

Fluxo Tpico:
1. Cliente envia DESCRIBE para obter informaes sobre o recurso (ex.: URL rtsp://exemplo.com/video).

2. Servidor responde com detalhes da mdia (formato, codecs).

3. Cliente envia SETUP para configurar o transporte (ex.: portas RTP/RTCP).

4. Cliente envia PLAY para iniciar o streaming.

5. Durante a transmisso, o cliente pode enviar PAUSE ou ajustar parmetros.

6. Ao final, TEARDOWN encerra a sesso.



```python
import cv2

cap = cv2.VideoCapture('http://honjin1.miemasu.net/nphMotionJpeg?Resolution=640x480&Quality=Standard')
#Ideal  o link rtsp da sua camera de video. Mas pode funcionar com uma fonte de video HTTP

while(1):
    ret, frame = cap.read()

    cv2.imshow('RTSP/HTTP transmissao', frame)

    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```
