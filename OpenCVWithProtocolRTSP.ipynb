# **RTSP Protocolo com OpenCV**

O RTSP (Real-Time Streaming Protocol) é um protocolo de rede utilizado para controlar a transmissão de conteúdo multimídia em tempo real, como áudio e vídeo. Ele foi desenvolvido para estabelecer e gerenciar sessões entre clientes (ex.: players de vídeo) e servidores de mídia (ex.: câmeras IP, servidores de streaming), permitindo operações como iniciar, pausar ou encerrar uma transmissão.

## Principais características:

Função Principal:
Controlar a reprodução de fluxos de mídia em tempo real, mas não transmite os dados em si. O transporte dos dados (áudio/vídeo) geralmente é feito por protocolos como RTP (Real-Time Transport Protocol) ou RTCP (Real-Time Control Protocol).

Métodos Similares ao HTTP:
- Usa comandos de texto, como:

    - OPTIONS: Consulta recursos suportados pelo servidor.

    - DESCRIBE: Obtém metadados sobre a mídia (ex.: formato, codec).

    - SETUP: Configura a conexão para transporte dos dados.

    - PLAY: Inicia a transmissão.

    - PAUSE: Pausa temporariamente.

    - TEARDOWN: Encerra a sessão.

Portas e Transporte:
- Geralmente opera na porta 554 (TCP ou UDP). O RTSP pode usar TCP para controle e UDP para transporte de mídia (priorizando baixa latência).

Fluxo Típico:
1. Cliente envia DESCRIBE para obter informações sobre o recurso (ex.: URL rtsp://exemplo.com/video).

2. Servidor responde com detalhes da mídia (formato, codecs).

3. Cliente envia SETUP para configurar o transporte (ex.: portas RTP/RTCP).

4. Cliente envia PLAY para iniciar o streaming.

5. Durante a transmissão, o cliente pode enviar PAUSE ou ajustar parâmetros.

6. Ao final, TEARDOWN encerra a sessão.



```python
import cv2

cap = cv2.VideoCapture('http://honjin1.miemasu.net/nphMotionJpeg?Resolution=640x480&Quality=Standard')
#Ideal é o link rtsp da sua camera de video. Mas pode funcionar com uma fonte de video HTTP

while(1):
    ret, frame = cap.read()

    cv2.imshow('RTSP/HTTP transmissao', frame)

    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```
